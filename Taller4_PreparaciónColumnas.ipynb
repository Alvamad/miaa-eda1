Para el caso de las valores de los exámenes médicos a-priori considero que utilizar el KNNImputer puede ser una buena opción al poder relacionar otras columnas como por ejemplo: Edad, Peso y Sexo.

Así como también utilizar otras variables donde sea pertinente, por e.j. para el colesterol total, los valores de colesterol HDL y LDL. O para los resultados del uroanálisis los resultados específicos (proteinas, cilindros, etc.)

```python
df['creatinine'] = imputer.fit_transform(df[['creatinine', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['glycated_hemoglobin'] = imputer.fit_transform(df[['glycated_hemoglobin', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['ldl_cholesterol'] = df['ldl_cholesterol'].str.replace(',', '.', regex=True)
df['ldl_cholesterol'] = imputer.fit_transform(df[['ldl_cholesterol','hdl_cholesterol', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['total_cholesterol'] = imputer.fit_transform(df[['total_cholesterol','ldl_cholesterol','hdl_cholesterol', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['glomerular_filtration_rate'] = imputer.fit_transform(df[['glomerular_filtration_rate', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['glucose'] = imputer.fit_transform(df[['glucose', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['triglycerides'] = df['triglycerides'].str.replace(',', '.', regex=True)
df['triglycerides'] = df['triglycerides'].str.replace('[^\d\.]', '', regex=True)  # Remove non-numeric characters

df['triglycerides'] = imputer.fit_transform(df[['triglycerides', 'age', 'weight', 'gender_encoded']])[:, 0]
```

# Taller 4. Preparación Columnas

En el taller anterior vimos cómo en general se prepararon algunas columnas para el análisis bivariado que se debía realizar.

Utilizamos **Label Encoding** para el género M/F y de esta manera poder utilizar la variable para el algoritmo del KNNImputer y poder rellenar los valores faltantes.

Así mismo también se utilizó en la variable objetivo para codificar el riesgo en un valor número con órden de magnitud y utilizarlo en la matriz de correlación.

Para este ejercicio, vamos a tomar la columna de Glucosa que es la que vemos que tiene más correlación con el riesgo, y la vamos a **discretizar** de acuerdo a los siguientes datos clínicos.

*   Menos de 100 mg/dL (5,6 mmol/L) se considera normal. (Se asignará 0)
*   Entre 100 y 125 mg/dL (5,6 a 6,9 mmol/L) se diagnostica como prediabetes. (Se asignará 1)
*   126 mg/dL (7,0 mmol/L) o más en dos pruebas distintas se diagnostica como diabetes. (Se asignará 2)

[Fuente](https://www.mayoclinic.org/es/diseases-conditions/prediabetes/diagnosis-treatment/drc-20355284#:~:text=L%20de%20sangre.-,En%20general%3A,distintas%20se%20diagnostica%20como%20diabetes.)





```python
df['glucose_discretized'] = pd.cut(df['glucose'], bins=[-float('inf'), 100, 125, float('inf')], labels=[0, 1, 2], right=False)
```

Pareciera que en términos de medicina, pudiésemos discretizar la mayoría de variables en relación a los valores de referencia de cada uno de los parámetros (bajo, normal, alto. Por e.j) y evaluar el impacto de cada transformación.

```python
correlation = df['glucose_discretized'].corr(df['cardiovascular_risk_encoded'])

print(f"Corr between glucose_discretized and cardiovascular_risk_encoded: {correlation}")
```

Observamos un incremento leve en la correlación: De 0.32 a 0.35 después de la discretización de la columna.

Probemos con una **normalización** y veamos si tiene otro impacto diferente. Vamos a utilizar el MinMaxScaler()

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df['glucose_normalized'] = scaler.fit_transform(df[['glucose']])

correlation = df['glucose_normalized'].corr(df['cardiovascular_risk_encoded'])

print(f"Corr between glucose_normalized and cardiovascular_risk_encoded: {correlation}")
```

En este caso no encontramos efecto en la correlación.

Pasemos ahora al caso de la perímetro abdominal, vamos a discretizar el perímetro de acuerdo al género y a los valores de referencia establecidos por la OMS.

Para Hombres

*   Menos de 94cm, normal. (Asigno 0)
*   Entre 94cm y 102cm, aumenta el riesgo. (Asigno 1)
*   Más de 102cm, riesgo elevado. (Asigno 2)


Para Mujeres

*   Menos de 80cm, normal. (Asigno 0)
*   Entre 80cm y 88cm, aumenta el riesgo. (Asigno 1)
*   Más de 88cm, riesgo elevado. (Asigno 2)

Observemos que esta discretización es un poco distinta o manual, porque dependemos de otra columna.

```python
def discretize_abdominal_circumference(row):
  if row['gender'] == 'M':
    if row['abdominal_circumference'] < 94:
      return 0
    elif 94 <= row['abdominal_circumference'] <= 102:
      return 1
    else:
      return 2
  elif row['gender'] == 'F':
    if row['abdominal_circumference'] < 80:
      return 0
    elif 80 <= row['abdominal_circumference'] <= 88:
      return 1
    else:
      return 2

df['abdominal_circumference_discretized'] = df.apply(discretize_abdominal_circumference, axis=1)

correlation = df['abdominal_circumference_discretized'].corr(df['cardiovascular_risk_encoded'])

print(f"Corr between abdominal_circumference_discretized and cardiovascular_risk_encoded: {correlation}")
```

En este caso, no obtenemos una mayor correlación con el riesgo cardiovascular al discretizar. 0.16 vs. 0.09

Veamos qué sucede al normalizar:

```python
scaler = MinMaxScaler()

df['abdominal_circumference_normalized'] = scaler.fit_transform(df[['abdominal_circumference']])

correlation = df['abdominal_circumference_normalized'].corr(df['cardiovascular_risk_encoded'])

print(f"Corr between abdominal_circumference_normalized and cardiovascular_risk_encoded: {correlation}")
```

Al igual que con la glucosa, al normalizar esta variable no obtenemos diferencias en la correlación.

## Conclusiones

1. El Label Encoding fue útil en el taller anterior para poder utilizar algoritmos de imputación y también para poder encontrar la correlación con los algoritmos existentes.

2. La discretización no siempre otorga resultados mayores a la hora de comparar la correlación con la variable objetivo. En el caso de la glucosa aumentó, pero en el perimetro abdominal disminuyó. Esto tal vez significa que los outliers de perímetro abdominal aportan más peso para su correlación con el riesgo. Es decir, es mejor no transformar esta columna, pero con la glucosa es mejor hacerlo al obtener mejores resultados.

3. La normalización no nos proporcionó en este caso diferencias con las columnas sin transformar. Puede indicar que para estos casos de datos médicos no es una transformación útil.
