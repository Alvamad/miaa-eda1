Procedemos a analizar una segunda variable que es el Diagnóstico de Diabetes, se encuentra en investigaciones médicas que la diabetes puede desencadenar un riesgo alto en pacientes, se comprueba si esto sucede en estos datos:

```python
df.diabetes.value_counts().plot(kind="bar")
plt.title("Diabetes")
plt.xlabel("Tiene Diabetes")
plt.xticks(rotation=0)
plt.ylabel("# Pacientes")
plt.show()
```

No es una muestra tan balanceada en cuanto a diagnóstico de diabetes, sin embargo es suficiente para análizar la relación con el riesgo.

```python
contingency_table = pd.crosstab(df['diabetes'], df['riesgo'])

plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.title('Correlation between Diabetes and Riesgo')
plt.xlabel('Riesgo')
plt.ylabel('Diabetes')
plt.show()
```

Como se observa en la tabla de contingencia, los pacientes diagnosticados con diabetes solo se encuentran en las categorías de riesgo Alto o Muy Alto, lo que nos confirma la información científica encontrada al respecto de la diabetes vs. riesgo de enfermedad cardiovascular.

```python
df['contraceptive_method'] = df.apply(lambda row: 'NO REGISTRA' if (pd.isna(row['contraceptive_method']) and row['gender'] == 'F')
                                     else ('NO APLICA' if row['gender'] == 'M' else row['contraceptive_method']), axis=1)

df.contraceptive_method.info()
```

Para el caso de las valores de los exámenes médicos a-priori considero que utilizar el KNNImputer puede ser una buena opción al poder relacionar otras columnas como por ejemplo: Edad, Peso y Sexo.

Así como también utilizar otras variables donde sea pertinente, por e.j. para el colesterol total, los valores de colesterol HDL y LDL. O para los resultados del uroanálisis los resultados específicos (proteinas, cilindros, etc.)

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['gender_encoded'] = label_encoder.fit_transform(df['gender'])

imputer = KNNImputer(n_neighbors=5)
```

```python
df['creatinine'] = imputer.fit_transform(df[['creatinine', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['glycated_hemoglobin'] = imputer.fit_transform(df[['glycated_hemoglobin', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['albuminuria'] = imputer.fit_transform(df[['albuminuria', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['urinary_creatinine'] = df['urinary_creatinine'].str.replace(',', '.', regex=True)  # Replace commas
df['urinary_creatinine'] = df['urinary_creatinine'].str.replace('[^\d\.]', '', regex=True)  # Remove non-numeric characters
df['urinary_creatinine'] = df['urinary_creatinine'].astype(float)  # Now convert to float

df['urinary_creatinine'] = imputer.fit_transform(df[['urinary_creatinine','creatinine', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['hdl_cholesterol'] = imputer.fit_transform(df[['hdl_cholesterol', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['ldl_cholesterol'] = df['ldl_cholesterol'].str.replace(',', '.', regex=True)
df['ldl_cholesterol'] = imputer.fit_transform(df[['ldl_cholesterol','hdl_cholesterol', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['total_cholesterol'] = imputer.fit_transform(df[['total_cholesterol','ldl_cholesterol','hdl_cholesterol', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['glomerular_filtration_rate'] = imputer.fit_transform(df[['glomerular_filtration_rate', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['glucose'] = imputer.fit_transform(df[['glucose', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['triglycerides'] = df['triglycerides'].str.replace(',', '.', regex=True)
df['triglycerides'] = df['triglycerides'].str.replace('[^\d\.]', '', regex=True)  # Remove non-numeric characters

df['triglycerides'] = imputer.fit_transform(df[['triglycerides', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['urinalysis_proteins'] = df['urinalysis_proteins'].replace({'neg': 0, 'pos': 1})

df['urinalysis_proteins'] = imputer.fit_transform(df[['urinalysis_proteins', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['urinalysis_rbc'] = df['urinalysis_rbc'].replace({'neg': 0, 'pos': 1})

df['urinalysis_rbc'] = imputer.fit_transform(df[['urinalysis_rbc', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['urinalysis_casts'] = df['urinalysis_casts'].replace({'neg': 0, 'pos': 1})

df['urinalysis_casts'] = imputer.fit_transform(df[['urinalysis_casts', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['urinalysis_result'] = df['urinalysis_result'].replace({'Negativo': 0, 'Positivo': 1})

df['urinalysis_result'] = imputer.fit_transform(df[['urinalysis_result','urinalysis_proteins','urinalysis_rbc','urinalysis_casts', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['previous_creatinine'] = imputer.fit_transform(df[['previous_creatinine', 'creatinine', 'age', 'weight', 'gender_encoded']])[:, 0]
```

```python
df['albuminuria_creatinine_ratio'] = df.apply(lambda row: round((row['albuminuria'] / row['creatinine']) * 100, 1)
                                                      if (pd.isna(row['albuminuria_creatinine_ratio']) and
                                                          not pd.isna(row['albuminuria']) and
                                                          not pd.isna(row['creatinine']))
                                                      else row['albuminuria_creatinine_ratio'], axis=1)
```
