```python
# Change names
df.rename(columns=
  {
      "CLASIFICACION FINAL DEL RIESGO": "riesgo",
      "Indice Masa Corporal = Peso/talla Al Cuadrado": "imc",
      "Perimetro abdominal": "perimetroAbdominal",
      "31_ColesterolTotal": "colesterolTotal",
      "32_HDL": "hdl",
      "EDAD (Años cumplidos)": "edad",
      "25_TenArtSis": "TASistolica",
      "20_DxDM": "diabetes",
      "82_Glicemia": "glicemia"
  }, inplace=True)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns

df.riesgo.value_counts().plot(kind="bar")
plt.title("Clasificación del Riesgo")
plt.xlabel("Tipo de riesgo")
plt.xticks(rotation=0)
plt.ylabel("# Pacientes")
plt.show()
```

Se analiza si la muestra está balanceada para la variable objetivo, se determina que está relativamente balanceada de manera favorable al tener más datos de pacientes con riesgo Alto-Muy Alto (sumados) y a su vez tener también datos de pacientes con riesgo Bajo-Moderado.

```python
df.perimetroAbdominal.plot(kind="box")
plt.title("Per. Abdominal")
plt.show()
```



Como se mencionó arriba, hay muchos outliers como se ve en este caso para la variable de perímetro abdominal.

```python
df.perimetroAbdominal.hist()
```

En este caso tenemos una distribución de datos relativamente normal, lo cual es beneficioso para el análisis.

```python
sns.catplot(x="riesgo", y="perimetroAbdominal", data=df, kind="box", aspect=1.5)
```

Acá podemos ver la primera apreciación, riesgo vs. perímetro abdominal. Se nota una "escalera" de riesgo bajo a muy alto entre tanto el promedio de perímetro abdominal aumenta. Esto puede significar estilos de vida sedentarios que conllevan a que el riesgo cardiovascular aumente.

De este notebook se excluyen los análisis de las demás variables al no encontrar nada concluyente (de manera inicial).

```python
df.diabetes.describe()
```

Procedemos a analizar una segunda variable que es el Diagnóstico de Diabetes, se encuentra en investigaciones médicas que la diabetes puede desencadenar un riesgo alto en pacientes, se comprueba si esto sucede en estos datos:

```python
df.diabetes.value_counts().plot(kind="bar")
plt.title("Diabetes")
plt.xlabel("Tiene Diabetes")
plt.xticks(rotation=0)
plt.ylabel("# Pacientes")
plt.show()
```

No es una muestra tan balanceada en cuanto a diagnóstico de diabetes, sin embargo es suficiente para análizar la relación con el riesgo.

```python
contingency_table = pd.crosstab(df['diabetes'], df['riesgo'])

plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.title('Correlation between Diabetes and Riesgo')
plt.xlabel('Riesgo')
plt.ylabel('Diabetes')
plt.show()
```

Como se observa en la tabla de contingencia, los pacientes diagnosticados con diabetes solo se encuentran en las categorías de riesgo Alto o Muy Alto, lo que nos confirma la información científica encontrada al respecto de la diabetes vs. riesgo de enfermedad cardiovascular.

```python
df.glicemia.hist()
```

Por último entonces revisamos la variable de los índices de glicemia de los pacientes, en la cual no se observa una dist. normal en los datos.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='riesgo', y='glicemia', data=df)
plt.title('Correlation between Glicemia and Riesgo')
plt.xlabel('Riesgo')
plt.ylabel('Glicemia')
plt.show()
```

También encontramos una gran cantida de outliers, pero con información relevante en cuanto al riesgo. Podemos observar la correlación entre el riesgo alto-muy alto con los índices de glicemia elevados, lo que por supuesto también correspondería (en su mayoría) con diagnósticos de diabetes.
